# Importing Libraries

```python
import pandas as pd
import numpy as np
```

# Importing Web Scraping libraries

```python
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
```

# Inputting the URL

```python
url= 'https://dejavu.shoes/collections/high-boots'
```

# Create a Client-based Request to Get the URL

```python
client = urlopen(url)
```

```python
#Getting the HTML Code of the Full Page

html = client.read()

html
```

```python
#Closing the Request

client.close()
```

# Creating an HTML Parser Using BeautifulSou

```python
soup = bs (html, "html.parser")

soup
```

# Creating a Container for the Needed Data

```python
containers = soup.find_all("div",{"class": "product-details"})

len(containers)
```

# Accessing Page Elements

```python
product = containers[0].findAll("a",{"class": "js-product-details-link"})

product[0].text.strip()
```

```python
price1 = containers[0].findAll("span",{"class": "price-item price-item--sale"})

price1[0].text.strip()
```

# Bringing it All Together

```python
f = open('dejavu.csv', 'w')
header = 'product,price_after_sale,price_before_sale\n'
f.write(header)

for container in containers:
    product_elements = container.findAll("a", {"class": "js-product-details-link"})
    price_after_sale_elements = container.findAll("span", {"class": "price-item price-item--sale"})
    price_before_sale_elements = container.findAll("s", {"class": "price-item price-item--regular"})

    if product_elements:
        product_name = product_elements[0].text.strip()
        price_after_sale = price_after_sale_elements[0].text.strip().replace(',', '.') if price_after_sale_elements else "N/A"
        price_before_sale = price_before_sale_elements[0].text.strip().replace(',', '.') if price_before_sale_elements else "N/A"

        f.write(f"{product_name},{price_after_sale},{price_before_sale}\n")
    else:
        # Handle the case when product is not found
        f.write("N/A,N/A,N/A\n")

f.close()
```
